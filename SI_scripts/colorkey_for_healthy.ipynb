```python
import matplotlib.patheffects as path_effects
import pickle 
import matplotlib.pyplot as plt 
import seaborn as sns   

## EDIT THE LINE FOLLOWING LINE ##
path_to_local_zenodo_download_file = 'some_path_to_your_local_copy_of_zenodo_download' 

all_of_interest_with_unclass = ['CAF1',
                                 'CAF2',
                                 'CAF3',
                                 'CAF4',
                                 'B Cell',
                                 'Cxcl2+ Cell',
                                 'Cytotoxic T Cell',
                                 'Dendritic Cell',
                                 'MAC1',
                                 'MAC2',
                                 'Regulatory T Cell',
                                 'Epithelial',
                                 'Unclassified']


# Define the order of cluster IDs
cluster_id_order = [
    'Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5', 'Cluster 6', 
    'Cluster 7', 'Cluster 8', 'Cluster 9', 'Cluster 10', 'Cluster 11', 'Cluster 12', 
    'Cluster 13', 'Cluster 14', 'Cluster 15', 'Cluster 16', 'Cluster 17', 'Cluster 18', 
    'Cluster 19']

# Load the cell ID to type mapping from the pickle file
cell_type_path = path_to_local_zenodo_download_file + '/misc_checkpoint_data/cell_id_to_type_dict_healthy.pkl'
with open(cell_type_path, 'rb') as f:
    cell_id_to_type_dict = pickle.load(f)

# Initialize an empty list to store colors for each cluster
cell_colors = []

# Assign colors to clusters based on their order
for i in range(len(cluster_id_order)):
    if i < 10:
        cell_colors.append(sns.color_palette('tab20')[(2 * i) % 20])
    else:
        cell_colors.append(sns.color_palette('tab20')[(2 * (i) - 10 + 1) % 20])

# Create a dictionary mapping each cluster ID to its corresponding color
new_colors = {j: cell_colors[cluster_id_order.index(j)] for j in cluster_id_order}



fig, ax = plt.subplots(1, 1, figsize=(2.5, 2.5))
for i, cell_type in enumerate(cluster_id_order):

    ax.text(i % 7, 0.4*(i // 7), cell_id_to_type_dict[cell_type], color=new_colors[cell_type], fontsize=45, ha='center', va='center')
ax.axis('off')

#fig.savefig('../temp_figures/figure_3/cell_type_legend.png', dpi=300, bbox_inches='tight')
```
